## Instalando Conda en Colab
---
Para obetener Ocean parcels actualizado y con todas sus dependencias se debe instalar bajo conda.

```python
# Descarga conda en Colab
%%bash
MINICONDA_INSTALLER_SCRIPT=Miniconda3-4.5.4-Linux-x86_64.sh
MINICONDA_PREFIX=/usr/local
wget https://repo.continuum.io/miniconda/$MINICONDA_INSTALLER_SCRIPT
chmod +x $MINICONDA_INSTALLER_SCRIPT
./$MINICONDA_INSTALLER_SCRIPT -b -f -p $MINICONDA_PREFIX
```

```python
#Instala y actualiza conda en colab
%%bash
conda install --channel defaults conda python=3.7 --yes
conda update --channel defaults --all --yes
```

```python
!conda --version
```

```python
!python --version
```

```python
#Esto coloca el la carpeta de los paquetes de colab en variable de entorno
import sys
_ = (sys.path
        .append("/usr/local/lib/python3.7/site-packages"))
sys.path
```

```python
# Esto crea en entorno o env de ocean parcels
# Que en realidad es el paquete mas todas sus dependencias (otros paquetes que necesita)
%%bash
conda create -n py3_parcels -c conda-forge parcels jupyter cartopy ffmpeg --yes
```

```python
#Esto Activa el env de ocean parcels
%%bash
source /usr/local/etc/profile.d/conda.sh
conda activate py3_parcels 
parcels_get_examples parcels_examples
```

```python
#Esto muestra que ambientes hay actualmente en conda
%%bash
conda info --envs
```

Finalemente, copiamos la carpeta del paquete y todas sus dependencias (otras carpetas de otros paquetes) a la ruta que tiene por defecto Colab para todos sus paquetes.
